# Different methods for stationarity

To make a time series stationary, there are several methods to stabilize the signal in the data. Use different time series to investigate the effect of the following methods on stationarity:

    Differencing
    Log Transform

For many datasets, it can be difficult to visually ascertain whether the time series is stationary or not just by plotting, so there is a statistical test called the **Augmented Dickey-Fuller test** that can help (`adfuller()` function in StatsModel).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.tsa.stattools import adfuller

%matplotlib inline
```

## Example of stationary distribution

```python
# Normal distribution

normal = pd.Series(np.random.randn(100))

normal.plot();
```

```python
# Gamma distribution

gamma = pd.Series(np.random.gamma(1, 1, 100))

gamma.plot();
```

## Airline passengers dataset

```python
airline = pd.read_csv('data/airline_passengers.csv', index_col='Month', parse_dates=True)
```

```python
airline.plot();
```

**Due to the upwards linear trend and seasonality, this time series is not stationary. Lets confirm with the ADF test:**

```python
adfuller(airline['Passengers'])
```

**The order of elements in the returned `adfuller` object are:**

    Test statistic
    P-value
    Number of time lags used
    Number of observations used
    Critical values for the test statistic at 1%, 5% and 10% levels
    Maximized Information Criterion
    
* **If the p-value is greater than 0.05, the time series is not stationary**
* **If the p-value is lower than 0.05, the time series is stationary**

**As you can see, the time series for airline passengers returns p-value of 0.99, which is far greater than 0.05 denoting non-stationarity.**

```python
# Function to run ADF test and print relevant results

def adf(x):
    res = adfuller(x)
    print("Test-Statistic:", res[0])
    print("P-Value:", res[1])
    
    if res[1] < 0.05:
        print("Stationary")
    else:
        print("Non-Stationary")
```

```python
# Test function

adf(airline['Passengers'])
```

```python
# Test function on normal distribution (note how small p-value is, i.e. very stationary data)

adf(np.random.randn(100))
```

```python
# Test function on gamma distribution

adf(np.random.gamma(1, 1, 100))
```

**What happens if we run the airline passengers time series through log transformation before running the ADF test?**

```python
airline['LogPassengers'] = np.log(airline['Passengers'])
```

```python
adf(airline['LogPassengers'])
```

```python
airline['LogPassengers'].plot();
```

**Ater log-transformation, even though the plot has stopped growing in variation, the distribution is still increasing steadily with seasonality hence still non-stationary.**

**What happens if we run the airline passengers time series through differencing before running the ADF test?**

```python
airline['Diff'] = airline['Passengers'].diff()
```

```python
airline['Diff'].plot();
```

```python
# Drop any null values that occurred when diff-ing - note how close p-value is to 0.05

adf(airline['Diff'].dropna())
```

**Differencing the airline passengers time series did not make it stationary, although note how close the p-value is to 0.05... The signal variance still increases over time although the upwards linear trend has been removed.**

**Lets try combining the two - log-transforming the diffed values - and see what happens:**

```python
airline['DiffLog'] = airline['LogPassengers'].diff()
```

```python
airline['DiffLog'].plot();
```

```python
adf(airline['DiffLog'].dropna())
```

**Differencing the log-transformed values still does not make the airline passenger time series stationary. Note how high the p-value is when compared to when the variance was increasing (after initial differencing). You may have to difference the data a second time:**

```python
airline['Diff_2Log'] = airline['DiffLog'].diff()
```

```python
adf(airline['Diff_2Log'].dropna())
```

```python
airline['Diff_2Log'].plot();
```

**Finally! The airline passengers time series is now stationary. In the ARIMA model, you should use log-transformed values with model parameter `d = 2`.**

## Stock prices dataset

**Remember to filter the dataset to a single pricing column for a single company, to make it a *univariate* time series, e.g. close stock prices for a company like Google or Starbucks.**

```python
stocks = pd.read_csv('data/sp500sub.csv', index_col='Date', parse_dates=True)
```

```python
# Google close stock prices

goog = stocks[stocks['Name'] == 'GOOG'][['Close']]
```

```python
goog.plot();
```

```python
# Log transformation and plot

goog['Log'] = np.log(goog['Close'])

goog['Log'].plot();
```

```python
# Differencing the log-transform and plot

goog['LogDiff'] = goog['Log'].diff()

goog['LogDiff'].plot();
```

**Visually the time series had a definite trend and after differencing the log-transformed values, it now looks stationary.**

**Lets run each series of values through the ADF test to confirm:**

```python
adf(goog['Log'])
```

```python
# Remember to drop null values from diffed results

adf(goog['LogDiff'].dropna())
```

**Nice! After differencing the log-transformed values, the time series is stationary. In fact, it is so stationary, the p-value went down to zero.**

**NOTE: Differencing the log-transformed values is the same as the log returns.**

```python
# Starbucks close stock prices

sbux = stocks[stocks['Name'] == 'SBUX'][['Close']]

# Log transformation
sbux['Log'] = np.log(sbux['Close'])

# Differencing the log-transform
sbux['LogDiff'] = sbux['Log'].diff()
```

```python
sbux['Log'].plot();
```

```python
sbux['LogDiff'].plot();
```

```python
adf(sbux['Log'])
```

```python
# Remember to drop null values from diffed results

adf(sbux['LogDiff'].dropna())
```

**Once again, differencing the log-transform values made the stock price time series stationary.**

**With these example datasets, it appears that differencing the data had more impact when enforcing stationarity in a time series.**
