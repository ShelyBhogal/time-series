# Auto ARIMA

**With the airline passengers dataset, build an Auto ARIMA model using the `auto_arima()` function available from pmdarimas library. It uses StatsModel under-the-hood, so the syntax should be familiar. It automatically finds the optimal model parameters for ARIMA or seasonal ARIMA model (SARIMAX) using a stepwise algorithm and AIC metric to minimize the loss, so you can use it like a forecasting model in itself. It also supports prediction methods.**

**When using ARIMA on the airline passengers data, you achieved pretty good results, but can auto ARIMA improve those results? Compare the auto ARIMA results to the custom ARIMA model from the previous session:**

    ARIMA(12, 1, 0) with log-transformed data, returned MSE score of 16.072

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import pmdarima as pm
```

```python
df = pd.read_csv('data/airline_passengers.csv', index_col='Month', parse_dates=True)
```

```python
df.head()
```

```python
# Log-transform the time series

df['LogPassengers'] = np.log(df['Passengers'])
```

```python
df.head()
```

```python
# Create training and test sets

N_test = 12

train = df.iloc[: -N_test]
test = df.iloc[-N_test :]
```

**The `auto_arima()` function creates the model, with some important parameters:**

* **`y` parameter contains the time series you are using. If there are other features, you can include them in parameter `X`.**
* **`d` parameter can be used to specify the number of times the data is differenced.  Note that when doing a full grid search (`stepwise=False`), the `d` and `D` parameters cannot be used, as they are automatically estimated by the function.**
* **`max_p` and `max_q` parameters are set quite low by default so you usually need to specify a maximum order for both AR and MA components that auto ARIMA must run through testing.**
* **`max_order` parameter can set a limit on the AR and MA components combined, i.e. the AR and MA orders are added together but cannot exceed the value specified in this argument. This promotes finding a simpler model.**
* **`trace` parameter prints out debugging information, i.e. each type of ARIMA model tested.**
* **`suppress_warnings` means that any warning messages coming from ARIMA will be squelched (StatsModel has lots of warnings).**
* **`seasonal` parameter flags a seasonal component that you have seen in the plots.**
* **If you enable the seasonal component, then you must specify the recorded periods of the time series in `m` parameter, e.g. 12 for monthly data, 7 for daily data or 52 for weekly data. If `m=1`, this means annual data, i.e. non-seasonal, so the `seasonal` flag will be turned off.**
* **By enabling the `stepwise` parameter, you ensure that the model uses stepwise algorithm to find the best combination of model parameters. Otherwise, the model uses full grid search to find the best combination.**



```python
# Train auto-arima model with non-logged data

model = pm.auto_arima(train['Passengers'], 
                      trace=True, 
                      suppress_warnings=True, 
                      seasonal=True, 
                      m=12)
```

```python
# Print results of stepwise search

model.summary()
```

**According to auto ARIMA, the best model is SARIMAX with `order=(3, 0, 0)`, which is the non-seasonal component showing as AR(3), and `seasonal_order=(0, 1, 0, 12)`, which performs one differencing only and goes back to 12 steps behind - all models will have the same seasonal period of 12. The model also has three weights and an intercept, with information showing whether any are statistically significant. At the bottom of the results, you also have the p-values for several statistical metrics.**

**The `auto_arima` function returns a model object off which you can make predictions from the test set using `predict()` method and specifying the number of timesteps to forecast in the future. When flagging confidence intervals to be generated, the function returns both the predictions and the confidence intervals, and both should be captured as separate variables for plotting.**

```python
test_pred, confint = model.predict(n_periods=N_test, return_conf_int=True)
```

```python
# Plot test data with predictions and confidence interval

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(test.index, test['Passengers'], label='data')
ax.plot(test.index, test_pred, label='forecast')
ax.fill_between(test.index, confint[:,0], confint[:,1], color='red', alpha=0.3)
ax.legend();
```

**The forecast is pretty good, with only two places where the values differ most (1960-03 and 1960-07). To see the full model performance, you can predict off the training set aswell, using `predict_in_sample()` method. Since you built the model on the training set, you only need to specify the start and end positions of an array, `0` to `-1`.**

```python
train_pred = model.predict_in_sample(start=0, end=-1)
```

```python
# Plot entire data with training and test predictions inc. confidence intervals

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df.index, df['Passengers'], label='data')
ax.plot(train.index, train_pred, label='fitted')
ax.plot(test.index, test_pred, label='forecast')
ax.fill_between(test.index, confint[:,0], confint[:,1], color='red', alpha=0.3)
plt.title("SARIMAX model with raw data")
ax.legend();
```

**As expected, the training predictions follow the training set closely. Note the flat line at the start, which happens when you add a seasonal component - the model cannot make predictions for the first few times, since you need values from a season behind, which do not exist.**

**Build the same auto ARIMA model with the log-transformed data instead, and compare the results, since we already know from the previous session that ARIMA prefers the log-transformed data:**

```python
logmodel = pm.auto_arima(train['LogPassengers'], 
                         trace=True, 
                         suppress_warnings=True, 
                         seasonal=True, 
                         m=12)
```

```python
logmodel.summary()
```

**Now we see that the best SARIMAX model is made up of non-seasonal component AR(2), and the seasonal component has one seasonal differencing with seasonal component MA(1)  - `SARIMAX order=(2, 0, 0) seasonal_order=(0, 1, 1, 12)`.**

**The model with raw data needed AR(3) non-seasonal component and one seasonal differencing only, so we can see with the log-transformed data the model becomes more 'parsimonious'.**

```python
# Plot test predictions with confidence interval

test_pred_log, confint = logmodel.predict(n_periods=N_test, return_conf_int=True)

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(test.index, test['LogPassengers'], label='data')
ax.plot(test.index, test_pred_log, label='forecast')
ax.fill_between(test.index, confint[:,0], confint[:,1], color='red', alpha=0.3)
ax.legend();
```

```python
# Plot entire data and predictions

train_pred_log = logmodel.predict_in_sample(start=0, end=-1)

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df.index, df['LogPassengers'], label='data')
ax.plot(train.index, train_pred_log, label='fitted')
ax.plot(test.index, test_pred_log, label='forecast')
ax.fill_between(test.index, confint[:,0], confint[:,1], color='red', alpha=0.3)
plt.title("SARIMAX model with log-transformed data")
ax.legend();
```

**Even though it looks like the log-transformed data performs better with the model, you can test quantitatively with RMSE metric.**

## Forecast RMSE

```python
def rmse(t, y):
    return np.sqrt(np.mean((t - y) ** 2))
```

```python
print("Non-logged RMSE:", rmse(test['Passengers'], test_pred))
print("Logged RMSE:", rmse(test['Passengers'], np.exp(test_pred_log)))
```

**Interesting! This is opposite to what you saw previously - it seems that auto ARIMA prefers the raw data, rather than the log-transformed data. Remember that the previous session with `ARIMA(12, 1, 0)` yielded RMSE of 16.072, which is close to the non-logged RMSE of 18.536 found here with `SARIMAX(order=(3, 0, 0), seasonal_order=(0, 1, 0, 12))`.**

## Build non-seasonal auto-arima with log-transformed data

**By suppressing the seasonal component, you can assess whether auto ARIMA would pick similar model order argument to the previous session with `ARIMA(12, 1, 0)`, or whether auto ARIMA can improve on those results. Make sure to use the log-transformed data to ensure fair comparison.**

**To ensure that the auto ARIMA even has a chance of finding the same model as yours, you need to set the `max_p`, `max_q` and `max_order` values for the non-seasonal AR and MA components.**

```python
model = pm.auto_arima(train['LogPassengers'],
                      trace=True,
                      suppress_warnings=True,
                      max_p=12, 
                      max_q=2, 
                      max_order=14,
                      stepwise=True,
                      seasonal=False)
```

```python
# It does not try that many values...order p for the AR component doesn't go above 5

model.summary()
```

**Interestingly, auto ARIMA very quickly estimated that `ARIMA(4, 1, 2)` are the best model parameters. This means AR(4) component with one differencing, and MA(2) component.**

```python
# Plot test predictions with confidence interval

test_preds, confints = model.predict(n_periods=N_test, return_conf_int=True)

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(test.index, test['LogPassengers'], label='data')
ax.plot(test.index, test_preds, label='forecast')
ax.fill_between(test.index, confints[:,0], confints[:,1], color='red', alpha=0.3)
ax.legend();
```

```python
train_preds = model.predict_in_sample(start=1, end=-1)

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df.index, df['LogPassengers'], label='data')
ax.plot(train.index[1:], train_preds, label='fitted')
ax.plot(test.index, test_preds, label='forecast')
ax.fill_between(test.index, confints[:,0], confints[:,1], color='red', alpha=0.3)
ax.legend();
```

```python
rmse(test['Passengers'], np.exp(test_pred))
```

**Yikes!! Auto ARIMA model (4, 1, 2) is not half as good as the custom ARIMA model (12, 1, 0). Just by the plots themselves, you can see that the model does not caputure the test data well, and the confidence interval is much larger. The RMSE value confirms the model performance deteriorating.**

## Build non-seasonal auto-arima with raw data

**You need to make sure that auto ARIMA cannot improve on the non-seasonal ARIMA results using the raw data, compared to the log-transformed data, since you got some interesting results when using auto ARIMA.**

**In this case, use full grid search rather than the stepwise algorithm - `stepwise=False` - to test all the possible combinations. This is not advised if you have seasonal and non-seasonal parts included.**

```python
model = pm.auto_arima(train['Passengers'],
                      trace=True,
                      suppress_warnings=True,
                      max_p=12, 
                      max_q=2, 
                      max_order=14,
                      stepwise=False,
                      seasonal=False)
```

```python
model.summary()
```

**Aha! Now we have similar to the previous ARIMA model session - with automated `ARIMA(12, 1, 2)` compared to custom `ARIMA(12, 1, 0)`. With auto ARIMA, there is an additional Moving Average MA(2) component. Lets assess the model with plot and RMSE value:**

```python
test_pred, confint = model.predict(n_periods=N_test, return_conf_int=True)

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(test.index, test['Passengers'], label='data')
ax.plot(test.index, test_pred, label='forecast')
ax.fill_between(test.index, confint[:,0], confint[:,1], color='red', alpha=0.3)
ax.legend();
```

```python
train_pred = model.predict_in_sample(start=1, end=-1)

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df.index, df['Passengers'], label='data')
ax.plot(train.index[1:], train_pred, label='fitted')
ax.plot(test.index, test_pred, label='forecast')
ax.fill_between(test.index, confint[:,0], confint[:,1], color='red', alpha=0.3)
plt.title("Auto ARIMA(12, 1, 2)")
ax.legend();
```

```python
rmse(test['Passengers'], test_pred)
```

**Still not as good as custom ARIMA(12, 1, 0) model but quite close... Furthermore, the custom ARIMA model has less parameters making it the better model.**

# Conclusion

**Auto ARIMA may not necessarily always find the best model parameters, especially with default parameter settings.**

**Minimizing the AIC does not mean that you are minimizing the RMSE, even though they are similar ideas in concept. The AIC cares more about model complexity than model accuracy, like the RMSE.**

**When stepwise algorithm is not testing many combinations, then it can be useful to use a full grid search, but watch out for the curse of dimensionality!**
