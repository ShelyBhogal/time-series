# ARIMA model

**Looking at the airline passengers dataset, build an ARIMA model to get an intuition of how it works under different `p`, `d` and `q` values, using StatsModel `ARIMA()` function.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('data/airline_passengers.csv', index_col='Month', parse_dates=True)
```

```python
df.head()
```

```python
df.plot(figsize=(15, 5));
```

**The time series increases steadily over time, with regular peaks and troughs indicating seasonality. This also shows that the time series is not stationary.**

**Visualize the effect of 'differencing' the time series with Pandas `diff()` function:**

```python
df['diff_1'] = df['Passengers'].diff()
```

```python
df.plot(figsize=(15, 5));
```

**The 'diffed' values are somewhat dampened but it is not quite stationary - there is still some increasing variation over time.** 

**Visualize the effect of log transform on the time series:**

```python
df['LogPassengers'] = np.log(df['Passengers'])
```

```python
df['LogPassengers'].plot(figsize=(15, 5));
```

**Notice that the log values are not growing as rapidly as the original time series, and the amplitude of the cycles look more constant than before, so log transform seems to make the time series more stationary than differencing.**

**Visualize the effect of differencing the log-transformed values:**

```python
df['Log1stDiff'] = df['LogPassengers'].diff()
```

```python
df['Log1stDiff'].plot(figsize=(15, 5));
```

**The behaviour of the time series is more what you want now - compared to the initial 'diffed' values, there is no growing variation over time, and compared to the 'non-diffed' log values, there is no upwards linear trend. The time series is now stationary!**

## Prepare time series for modeling

```python
df.index.freq = 'MS'

Ntest = 12

train = df.iloc[:-Ntest]
test = df.iloc[-Ntest:]
```

```python
train_idx = df.index <= train.index[-1]

test_idx = df.index > train.index[-1]
```

## Build ARIMA (1, 0, 0) model

**To start with, build an ARIMA model where the current value depends *linearly* on the value immediately prior. This is equivalent to the model order `(1, 0, 0)`, i.e. lag of one previous timestep, zero differencing, zero lags of past errors.**

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
# order of model (p, d, q)

arima = ARIMA(train['Passengers'], order=(1, 0, 0))
```

```python
arima_result = arima.fit()
```

```python
# Generate training predictions and add to DataFrame

df.loc[train_idx, 'AR(1)'] = arima_result.predict(start=train.index[0], end=train.index[-1])
```

```python
df[['Passengers', 'AR(1)']].plot(figsize=(15, 5));
```

**The training predictions follow the original time series closely but lagging behind by one month (similar behaviour to Holt-Winters), which is expected since we are forecasting.**

**Note the initial predicted value, i.e. when x=0, is at least three times the original value.**

```python
# Generate test predictions and addd to DataFrame

prediction_result = arima_result.get_forecast(Ntest)

forecast = prediction_result.predicted_mean

df.loc[test_idx, 'AR(1)'] = forecast
```

```python
df[['Passengers', 'AR(1)']].plot(figsize=(15, 5));
```

**As you can see, this model is not very good. The training predictions are made by taking the last value, which is no better than naive forecasting, and the test forecasts do not even capture the upwards linear trend - instead it goes off in a completely opposite direction. So naive forecasting would have been better on the test set than the model.**

**Lets explore the forecast object more:**

```python
prediction_result 
```

```python
# Confidence levels

prediction_result.conf_int()
```

```python
# Function to plot training and test predictions with confidence levels

def plot_fit_and_forecast(result):
    fig, ax = plt.subplots(figsize=(15, 5)) 
    ax.plot(df['Passengers'], label='True data')
    
    # Plot the curve fitted on train set 
    train_pred = result.fittedvalues
    ax.plot(train.index, train_pred, color='green', label='Fitted data')
    
    # Forecast the test set
    prediction_result = result.get_forecast(Ntest)
    conf_int = prediction_result.conf_int()
    lower, upper = conf_int['lower Passengers'], conf_int['upper Passengers']
    forecast = prediction_result.predicted_mean
    ax.plot(test.index, forecast, label='Forecast data')
    ax.fill_between(test.index, lower, upper, color='red', alpha=0.3)
    ax.legend()
```

```python
plot_fit_and_forecast(arima_result)
```

## Build ARIMA (10, 0, 0) model

**In the case when your model is performing badly, what is the next course of action? Add more inputs! Build a new ARIMA model of order `(10, 0, 0)`, i.e. AR(10) model with zero differencing and zero past errors.**

```python
arima = ARIMA(train['Passengers'], order=(10, 0, 0)) # AR(10)

arima_result = arima.fit()

# Plot predictions with confidence levels
plot_fit_and_forecast(arima_result)
```

**Regardless of warning error message, the training and test predictions fit the data much better, by capturing the linear trend and seasonality. Additionally, there is no 'lagging' characterisitic, so the model is learning to anticipate the pattern. However, the increasing amplitude has not been captured at all, in fact it has reduced.**

## Build ARIMA (0, 0, 1) model

**Lets see how the Moving Average component influences the model behaviour, by building a pure MA ARIMA model of order `(0, 0, 1)`, i.e. zero past values and zero differencing, only the immediately previous error is used to predict.**

```python
arima = ARIMA(train['Passengers'], order=(0, 0, 1)) # MA(1)

arima_result = arima.fit()

plot_fit_and_forecast(arima_result)
```

**Ew. Worst model by far. The moving average expected forecast is always a constant value (see the straight prediction line), so intuitively, you should know this was the wrong choice.**

## Build ARIMA (8, 1, 1) model

**Earlier, you differenced the log transformations, which made the time series stationary. You knew at that point that the data needs to be differenced at least once, so make sure to add differencing to the model parameters.**

```python
# First build new model without log-transformed data

arima = ARIMA(train['Passengers'], order=(8, 1, 1))

arima_result_811 = arima.fit()
```

```python
# Function to plot predictions and confidence levels
# New function needed because differencing removes first row - d argument represents number of rows 

def plot_fit_and_forecast_int(result, d, col='Passengers'):
    fig, ax = plt.subplots(figsize=(15, 5))
    ax.plot(df[col], label='True data')
    
    # Plot the curve fitted on train set
    train_pred = result.predict(start=train.index[d], end=train.index[-1])
    
    ax.plot(train.index[d:], train_pred, color='green', label='Fitted data')
    
    # Forecast the test set
    prediction_result = result.get_forecast(Ntest)
    conf_int = prediction_result.conf_int()
    lower, upper = conf_int[f'lower {col}'], conf_int[f'upper {col}']
    forecast = prediction_result.predicted_mean
    ax.plot(test.index, forecast, label='Forecast data')
    ax.fill_between(test.index, lower, upper, color='red', alpha=0.3)
    ax.legend()
```

```python
plot_fit_and_forecast_int(arima_result_811, 1)
```

**As you can see, this does better than the pure AR model and the forecast nicely captures the seasonality. However, this model still underestimates the volume.**

## Re-build ARIMA model with log-transformed data

**Keeping the same model order as before `(8, 1, 1)`, use the log-transformed data to train the ARIMA model.**

```python
arima = ARIMA(train['LogPassengers'], order=(8, 1, 1))

arima_result_log811 = arima.fit()

plot_fit_and_forecast_int(arima_result_log811, 1, col='LogPassengers')
```

**With log-transformed data, the model performs fairly well, but not as well as using non-transformed data and still does not capture the increase in volume.**

## Build ARIMA (12, 1, 0) model

**What happens if you use a pure AR model and increase the number of previous time lags to match the length of the year (last 12 months of data), i.e. `(12, 1, 0)` with zero influence from previous error terms.**

**NOTE: First try with the non-transformed data.**

```python
arima = ARIMA(train['Passengers'], order=(12, 1, 0))

arima_result_1210 = arima.fit()

plot_fit_and_forecast_int(arima_result_1210, 1, col='Passengers')
```

**Better! Now lets see what happens if you use the same model parameters on the log-transformed data:**

```python
arima = ARIMA(train['LogPassengers'], order=(12, 1, 0))

arima_result_log1210 = arima.fit()

plot_fit_and_forecast_int(arima_result_log1210, 1, col='LogPassengers')
```

**Best so far!! However, as a scientist you should know that visual confirmation is not enough. You should run performance metrics to properly assess which model is best.**

## Performance metrics

```python
# Function to calculate RMSE 

def rmse(result, is_logged):
    forecast = result.forecast(Ntest)
    
    if is_logged:
        forecast = np.exp(forecast)
        
    t = test['Passengers'] 
    y = forecast 
    return np.sqrt(np.mean((t - y)**2))
```

```python
print("ARIMA(8, 1, 1):", rmse(arima_result_811, False))
print("ARIMA(8, 1, 1) logged:", rmse(arima_result_log811, True))
print("ARIMA(12, 1, 0):", rmse(arima_result_1210, False))
print("ARIMA(12, 1, 0) logged:", rmse(arima_result_log1210, True))
```

**Now you can confirm that the ARIMA (12, 1, 0) model with log-transformed data best captures the airline passengers dataset, because it returns the lowest RMSE score.**
