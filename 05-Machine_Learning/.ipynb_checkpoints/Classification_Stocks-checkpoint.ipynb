# Classification models with stock prices dataset

**Using the same stock data as the last exercise, apply traditional classification models to a univariate time series. Instead of predicting a price (or return), you are aiming to predict the 'direction' of the price, i.e. are prices going up or down. This should make the task easier than predicting numerical value, which has an infinite number of possibilities.**

**In this exercise, only predict one step ahead, i.e. one-step forecast, which the SciKit-Learn classification models do automatically.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
df0 = pd.read_csv('data/sp500sub.csv', index_col='Date', parse_dates=True)
```

```python
df = df0[df0['Name'] == 'IBM'][['Close']].copy()
```

```python
df['LogClose'] = np.log(df['Close'])
```

```python
df['LogReturn'] = df['LogClose'].diff()
```

```python
# Differencing creates initial null value

df.head()
```

```python
df[-100:].plot(figsize=(15, 5));
```

```python
# Obtain test set size
N_test = 252

# Strictly speaking this is not necessary since training and test sets are created from X and Y
#train = df.iloc[:-N_test]
#test = df.iloc[-N_test:]
```

## Make 'supervised' dataset, i.e. X and Y

**The stationary time series is the input *X*, and the target output *Y* is simply whether the log return is negative or positive, i.e. less than zero or greater than zero respectively. Compared to the regression tasks, the target is pre-computed, rather than coming from the time series itself.**

**NOTE: The higher the *T* value (number of past lags), the more slack you have in predicting values.**

```python
# Make supervised dataset - use T past values to predict the next value

series = df['LogReturn'].to_numpy()[1:] # First value is nan
target = (series > 0) * 1

T = 21
X = []
Y = []

# Count up to 1 less than before, since there is no target for last close
for t in range(len(series) - T):
    x = series[t:t+T]
    X.append(x)
    y = target[t+T]
    Y.append(y)

# Convert to NumPy arrays
X = np.array(X).reshape(-1, T)
Y = np.array(Y)

N = len(X)

print("X.shape", X.shape, "Y.shape", Y.shape)
```

```python
# Split X and Y into X_train, y_train, X_test, y_test

X_train, y_train = X[:-N_test], Y[:-N_test]

X_test, y_test = X[-N_test:], Y[-N_test:]
```

## Logistic Regression

```python
lr = LogisticRegression()

lr.fit(X_train, y_train)

print("Train Accuracy score:", lr.score(X_train, y_train))
```

```python
print("Test Accuracy score:", lr.score(X_test, y_test))
```

**Both training and test accuracy scores for Logistic Regression are approx. 50%, which is no better than predicting by tossing a coin.**

## Support Vector Machine

```python
svc = SVC()

svc.fit(X_train, y_train)

print("Train Accuracy score:", svc.score(X_train, y_train))
```

```python
print("Test Accuracy score:", svc.score(X_test, y_test))
```

**The training score has improved but the test score is 50% again, which is not good. You can conclude that the model overfits.**

## Random Forests

```python
rf = RandomForestClassifier()

rf.fit(X_train, y_train)

print("Train Accuracy score:", rf.score(X_train, y_train))
```

```python
print("Test Accuracy score:", rf.score(X_test, y_test))
```

**Unsurprisingly, the training accuracy score is perfect, but the test score is just below 50% again, which means the model is overfitting.**

**In conclusion, the linear Logistic Regression is the only model that did not overfit.**
